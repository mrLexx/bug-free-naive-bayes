```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import matplotlib.mlab as mlab
from scipy.stats import norm


x = np.array ([85, 90, 70, 95, 91])# Заполните здесь свой список данных. Если он уже находится в формате массива, нет необходимости конвертировать
#n, bin, patches = plt.hist (x, 20, плотность = 1, facecolor = 'blue', alpha = 0,75) # Второй параметр - это количество столбцов гистограммы
mu = np.mean (x) # Рассчитать среднее 
sigma =np.std(x) 
num_bins = 30 # Количество столбцов гистограммы 
n, bins, patches = plt.hist(x, num_bins,density=1, alpha=0.75) 
# Функция гистограммы, x - это значение оси x, normed = 1 - плотность вероятности, то есть сумма равна единице, зеленый квадрат, параметр глубины цвета равен 0.5. Возвращает n вероятностей, значение x левой линии прямого квадрата и каждый квадратный объект 
y = norm.pdf (bins, mu, sigma) # Подходит для лучшей кривой нормального распределения y 

plt.grid(True)
plt.plot (bins, y, 'r--') # Рисовать кривую y 
plt.xlabel ('values') # Рисовать ось x 
plt.ylabel ('Вероятность') # Рисовать ось у 
plt.title ('Гистограмма: $ \ mu $ =' + str (round (mu, 2)) + '$ \ sigma = $' + str (round (sigma, 2))) # Китайский заголовок u'xxx ' 
# plt.subplots_adjust (left = 0.15) # Оставленный запас 

plt.show()
```
